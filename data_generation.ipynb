```python
import random

# Function to load data from a text file into a list
def load_data(filename):
    with open(filename, 'r') as file:
        return [line.strip() for line in file.readlines()]

# Load data
programming_languages = load_data('data/pl.txt')
software = load_data('data/sw.txt')
roles = load_data('data/ro.txt')
skills = load_data('data/sk.txt')
certifications = load_data('data/cr.txt')

# Generate a job description with random selections
def generate_job_description():
    
    mapping = {
        "RO":random.choice(roles),
        ""
    }
    template = f"""
    Title: {}
    Responsibilities:
    - Develop applications using {random.choice(programming_languages)} and {random.choice(programming_languages)}
    - Manage database solutions with {random.choice(software)}

    Required Skills:
    - Proficient in {random.choice(skills)}
    - Experience with {random.choice(software)}

    Preferred Certifications:
    - {random.choice(certifications)}

    Benefits:
    - Competitive salary
    - Health benefits
    """
    return template

# Tagging function for entities with B-I notation
def tag_description(description, entities, tag_prefix):
    tokens = description.split()
    tagged_tokens = []
    for token in tokens:
        token_tagged = False
        for entity in entities:
            entity_words = entity.split()
            if token in entity_words:
                index = entity_words.index(token)
                if index == 0:
                    tagged_tokens.append(f"{token}/B-{tag_prefix}")
                else:
                    tagged_tokens.append(f"{token}/I-{tag_prefix}")
                token_tagged = True
                break
        if not token_tagged:
            tagged_tokens.append(f"{token}/O")
    return ' '.join(tagged_tokens)

# Generate and print a job description
job_description = generate_job_description()
print(job_description)

# Tag the job description
tagged_job_description = tag_description(job_description, programming_languages, 'PL')
tagged_job_description = tag_description(tagged_job_description, software, 'SW')
tagged_job_description = tag_description(tagged_job_description, roles, 'RO')
tagged_job_description = tag_description(tagged_job_description, skills, 'SK')
tagged_job_description = tag_description(tagged_job_description, certifications, 'CR')
print(tagged_job_description)
```

```python
from itertools import combinations
import os
import random

class Keywords():
    
    def __init__(self,path):
        
        self.path = path
        self.tags_to_idx = {}
        self.idx_to_tags = {}

        self.cm = []
        self.ro = []
        self.sk =[]

        self.CM = "cm"
        self.RO = "ro"
        self.SK = "sk"
        self.history = set()
        self.load_data()

        
    def load_file(self,file_path):
        
        with open(file_path, 'r', encoding='utf-8') as file:
            return [line.strip() for line in file.readlines()]
        

    def load_data(self):
        
        files = os.listdir(self.path)
        

        cnt=0
        for i in range(len(files)):
            if files[i][:2]!='jd':
                self.tags_to_idx[files[i][:2]]=cnt
                self.idx_to_tags[cnt]=files[i][:2]
                cnt+=1
                
        for file in files:
            
            file_path = os.path.join(self.path,file)
            
            keywords = self.load_file(file_path)

            if file[:2]==self.SK:
                self.sk = keywords
            elif file[:2]==self.RO:
                self.ro = keywords
            elif file[:2] ==self.CM:
                self.cm = keywords
        print("Loaded Keywords!")
        
    def get_keywords(self,no_sk,no_ro):

        return {

            self.SK:self.get_combinations(self.SK,no_sk),
            self.RO:self.get_combinations(self.RO,no_ro),
            self.CM:self.get_combinations(self.CM,1)
        }


    def get_combinations(self,category ,n):
        
        if category==self.SK:
            input_list = self.sk
        elif category==self.RO:
            input_list = self.ro
        elif category ==self.CM:
            input_list = self.cm
        
        cnt=0
        while True:
            
            combination = tuple(random.sample(input_list,n))
            cnt+=1
            if combination  not in self.history or cnt>700:
                break
        self.history.add(combination)

        
        return combination
    
    def clear_history(self):
        
        self.history = set()
    
    
curr_dir = os.getcwd()
path = os.path.join(curr_dir,"data")
    
k = Keywords(path)

# print(k.load_file(os.path.join(path,"sk.txt")))

print(k.get_keywords(10,3))
```

```python
import os
import nltk
# nltk.download("punkt")
from nltk.tokenize import word_tokenize
import pandas as pd

curr_dir = os.getcwd()
keywords_path = os.path.join(curr_dir,"data")
    
keywords = Keywords(keywords_path)

class DataItem():
    
    def __init__(self,tokens,tag_index):
        
        self.tokens = tokens
        self.tag_index= tag_index

        self.CM = "cm"
        self.RO = "ro"
        self.SK = "sk"
        self.tags = set([self.SK,self.RO,self.CM])

        
#         self.get_all()
        
    def generate_data(self,id):
        
   

        no_sk=len(self.tag_index[self.SK])
        no_ro=len(self.tag_index[self.RO])
        no_cm=len(self.tag_index[self.CM])

        counter = {
            self.RO:0,self.SK:0,self.CM:0
        }
        
        data = keywords.get_keywords(no_sk=no_sk,no_ro=no_ro)
        
        final_data = []
        for i in range(len(self.tokens)):
            tag = self.tokens[i][1:3]
            if tag in self.tags:
                company = data[tag][min(len(data[tag])-1,counter[tag])].split()
                counter[tag]+=1
                for i,word in enumerate(company):
                    if i==0:
                        tag_ = "B"
                    else:
                        tag_ ="I"
                    final_data.append((id,word,tag_+"-"+tag))
            
            else:
                final_data.append((id,self.tokens[i],"O"))

        return final_data
            


class NerDatasetCustom():
    
    def __init__(self,path_data,size,config):
        
        self.path_data = path_data
        self.size = size
        self.config = config
        self.jds = []
        self.len_jds = 0
        self.keywords_to_idx = {}
        self.idx_to_keywords = {}
        self.tags_to_idx = {}
        self.idx_to_tags = {}
        self.dataset = []
        self.clear_history_every = 5000
        self.load_data()
        # self.create_dataset()
    
    def load_file(self,file_path):
        
        with open(file_path, 'r', encoding='utf-8') as file:
            return [line.strip() for line in file.readlines()]

        
    def load_data(self):
        
        files = os.listdir(self.path_data)
        
        # print(self.tags_to_idx)
        for file in files:
            
            if file[:2]=='jd':
                
                t_path = os.path.join(self.path_data,file)
                data = self.load_file(t_path)
                data = [self.clean_text(text) for text in data]
                data = " ".join(data)
                tokens = word_tokenize(data)
                filtered_tokens = []
                word_t=""

                tag_index = {}
                
                for token in tokens:
#                     print(token)
                    if token=="<":
                        word_t+=token
                    elif token==">":
                        word_t+=token
                        if word_t[1:3] not in tag_index.keys():
                            tag_index[word_t[1:3]]=[]
                        tag_index[word_t[1:3]].append(len(filtered_tokens))
                        filtered_tokens.append(word_t)
                        word_t=""
                    elif len(word_t)!=0:
                        word_t+=token
                    else:
                        filtered_tokens.append(token)

                self.jds.append(DataItem(filtered_tokens,tag_index))
        self.len_jds = len(self.jds)
        print("Successfully Loaded all Text Files!")


    def create_dataset(self):
                # print(tag_index)
        print("Generating DataFrame!")
        for i in range(self.size):
            jd_idx = random.randint(0,self.len_jds-1)
            self.dataset.append(self.jds[jd_idx].generate_data(i))
            if i%self.clear_history_every==0:
                keywords.clear_history()
        
        return self.create_dataframe(self.dataset)
    
    def create_dataframe(self,data):
    # Flatten the list of lists and create a DataFrame
        flattened_data = [item for sublist in data for item in sublist]
        df = pd.DataFrame(flattened_data, columns=['id','word', 'tag'])
        print(f"Successfully Created DataFrame with {self.size} sentences.")
        return df
                
    def clean_text(self,text):

        # Remove unwanted characters
        characters_to_remove = ['*', '-',';','[',']']
        special_ch = "/"
        for char in characters_to_remove:
            text = text.replace(char, '')
        text = text.replace(special_ch,'')

        # Convert to lowercase
        # text = text.lower()

        return text
        
                
curr_dir = os.getcwd()
data_path = os.path.join(curr_dir,"data")

ner_data = NerDatasetCustom(data_path,30000,{})

dataset = ner_data.create_dataset()
        
```
